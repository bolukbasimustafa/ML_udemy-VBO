## Statsmodels ile modelleme

```python
import statsmodels.api as sm
```

```python
X = df[["TV"]]     # basit doÄŸrusal regresyon modelinde tek deÄŸiÅŸken olarak tv' yi X baÄŸÄ±msÄ±z deÄŸiÅŸkeni olarak alÄ±yoruz
X[0:5]
```

```python
X = sm.add_constant(X)    # tek deÄŸiÅŸkenli bir matrise 1 sabiti ekleyerek iki deÄŸiÅŸkenli matris yaparak matris 
                          # iÅŸlemlerinin yapÄ±labilmesini saÄŸlÄ±yoruz. aÅŸaÄŸÄ±daki matrise bakarak durumu gÃ¶relim
```

```python
X[0:5]
```

```python
model = lm.fit()               # gercek model fit ediliyor
```

```python
model.summary()                

# modelin Ã§Ä±ktÄ±larÄ±na ulaÅŸÄ±lÄ±yor. burdaki geniÅŸ Ã§aplÄ± Ã§Ä±ktÄ± ve istatistiksel testleri sklearn ile alamÄ±yoruz, 
# bu nedenle bu tip doÄŸrusal regresyon modelinde istatistiksel anlamdaki detaylÄ± deÄŸerlere eriÅŸmek iÃ§in statsmodels
# fonksiyonu kÃ¼tÃ¼phanesi ile gerÃ§ekleÅŸtirebiliyoruz
```

* R-squared: Ã§Ä±ktÄ±nÄ±n en Ã¶nemli yanlarÄ±ndan biridir. baÄŸÄ±msÄ±z deÄŸiÅŸkenin baÄŸÄ±mlÄ± deÄŸiÅŸkendeki deÄŸiÅŸkenliÄŸi aÃ§Ä±klama baÅŸarÄ±sÄ±dÄ±r. Elimizdeki TV deÄŸiÅŸkeni sales deÄŸiÅŸkeninde yer alan deÄŸiÅŸkenliÄŸin yaklaÅŸÄ±k %60 (0.612) Ä±nÄ± aÃ§Ä±klayabilmektedir.

* Adj. R-squared (DÃ¼zeltilmiÅŸ R-squared) : R-squared ile Adj. R-squared arasÄ±ndaki fark ÅŸudur: R-squared fromulasyonu itibariyle modele ne kadar fazla deÄŸiÅŸken eklersek ekleyelim ÅŸiÅŸmeye meyillidir artacaktÄ±r. DolayÄ±sÄ±yla modelde ilgili ilgisiz anlamlÄ± anlamsÄ±z deÄŸiÅŸken olmasÄ± durumunda artacaktÄ±r. R-squared aÃ§Ä±klanabilirlik oranÄ±dÄ±r.  DolayÄ±sÄ±yla modelde ilgili ilgisiz anlamlÄ± anlamsÄ±z deÄŸiÅŸken ekleyip R-squared in arttÄ±ÄŸÄ±nÄ± gÃ¶rdÃ¼ÄŸÃ¼mÃ¼zde  iyi bir yolda olduÄŸumuzu dÃ¼ÅŸÃ¼nebilirz. Bu durum Ã§ok yanÄ±ltÄ±cÄ± bir durumdur.  Adj. R-squared; R-squared in her parametre eklenmesine karÅŸÄ± duyarlÄ±lÄ±ÄŸÄ± tÃ¶rpÃ¼leyen dÃ¼zenleyen duyarlÄ±lÄ±ÄŸÄ± daha az olan bir metriktir. Birden fazla deÄŸiÅŸken olduÄŸunda Adj. R-squared deÄŸeri bizim iÃ§in daha deÄŸerli ve dÃ¼zgÃ¼n bir deÄŸerdir. 

* F-statistic: Modelin anlamlÄ±lÄ±ÄŸÄ±nÄ±n anlaÅŸÄ±lmasÄ± iÃ§in kurulan test istatistiÄŸidir. Kurulan model bize bazÄ± tahminler verecek ve  belli bir tahmin baÅŸarÄ±sÄ± olacaktÄ±r ama bu model anlamlÄ±mÄ± sorsunu istatistiki olarak test edip modelin anlamlÄ±lÄ±ÄŸÄ±na yÃ¶nelik yapÄ±lan F istatistiÄŸidir.

* Prob (F-statistic): Modelin anlamlÄ±lÄ±ÄŸÄ±na yÃ¶nelik yapÄ±lan F istatistiÄŸinin P-value deÄŸeridir. P-value ve F-statistic birlikte incelenerek anlamlÄ±lÄ±ÄŸÄ± incelenir.  

* AIC ve BIC : Model karÅŸÄ±laÅŸtÄ±rma iÃ§in kullanÄ±lan 2 istatistik deÄŸeridir. ÅŸuan ilgilenmiyoruz

* No. Observations : GÃ¶zlem sayÄ±sÄ± | Df Residuals : Serbestlik derecesi bunlarlada ilgilenmiyoruz

* modelin en anlamlÄ± kÄ±smÄ±dÄ±r.

* const : Beta0, TV : Beta1  deÄŸeri olan sabit deÄŸiÅŸkenlerdir. coef parametre olarak ifade ettiÄŸimiz katsayÄ± deÄŸeridir. KatsayÄ±larÄ±n anlamlÄ± olup olmadÄ±klarÄ±nÄ± anlamak iÃ§in t istatistiÄŸi ve t istatistiÄŸine iliÅŸkin P-value (P>|t|) deÄŸerleri verilmiÅŸ. Burdan anlamlÄ± olduÄŸu anlaÅŸÄ±ÅŸÄ±lÄ±yormuÅŸ ama nasÄ±l Ã¶ÄŸrenmek lazÄ±m. 0.025 ve 0.975 ilede katsayÄ±larÄ±n gÃ¼ven aralÄ±ÄŸÄ± verilmiÅŸ. 

* burdan ÅŸÃ¶yle bir sonuc Ã§Ä±kÄ±yor:  TV deÄŸiÅŸkenine ait 0.0475 katsayÄ± deÄŸeri (t istatistiÄŸi ve buna baÄŸlÄ± P-value deÄŸerine gÃ¶re) istatiksel olarak anlamlÄ±dÄ±r ve bu katsayÄ± deÄŸeri yine istatiksel olarak %95 gÃ¼venilirlikle 0.042 ve 0.053 aralÄ±ÄŸÄ±nda yer alacaktÄ±r. Bu yorum deÄŸerli bir bilgidir. Buna bakarak bazÄ± stratejiler belirlenebilir

* TV nin katsayÄ± deÄŸeri olan 0.0475 ise TV reklam harcamalaraÄ±nda 1 br lik artÄ±ÅŸ olduÄŸunda sales de ortalama beklenen artÄ±ÅŸ miktarÄ±dÄ±r olarak yorumlanÄ±r.

### model kurma iÃ§in ikinci yol

```python
import statsmodels.formula.api as smf
lm = smf.ols("sales ~ TV", df)          # baÄŸÄ±mlÄ± ve baÄŸÄ±msÄ±z deÄŸiÅŸken isimleri ile yazÄ±lÄ±yor
model = lm.fit()
model.summary()
```

```python
model.params           # modelin parametre deÄŸerleri. Beta0 ve Beta1 deÄŸerleri
```

```python
model.summary().tables[1]          # katsayÄ±larla iliÅŸkin deÄŸerlerin olduÄŸu tables Ä±n ilgili kÄ±smÄ±nÄ± veriyor
```

```python
model.conf_int()                   # katsayÄ±larÄ±n gÃ¼ven aralÄ±ÄŸÄ±. yukarÄ±daki kÄ±smÄ±n 0.025 ile 0.975 aralÄ±ÄŸÄ± kÄ±smÄ±nÄ± alÄ±r 
                                   # yani %95 gÃ¼ven aralÄ±ÄŸÄ±
```

```python
model.f_pvalue                    # sadece modelin anlamlÄ±lÄ±ÄŸÄ±na iliÅŸkin istatistiklere pvalue ya ulaÅŸmak iÃ§in
```

```python
print("f_pvalue: ", "%.4f" % model.f_pvalue)        # p-value deÄŸerini e den kurtarmak iÃ§in
```

```python
print("fvalue: ", "%.2f" % model.fvalue)            # f-statistic deÄŸerine ulaÅŸmak iÃ§in
```

```python
print("tvalue: ", "%.2f" % model.tvalues[0:1])      # parametre anlamlÄ±lÄ±ÄŸÄ±na iliÅŸkin deÄŸerlere ulaÅŸmak iÃ§in 
```

```python
model.mse_model      # (hata kareler ortalamasÄ±) modelin anlamlÄ±lÄ±ÄŸÄ±na iliÅŸkin model deÄŸerlendirme istatistiklerine ulaÅŸmak iÃ§in
```

## Model Tuning / Model DoÄŸrulama

```python
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df["sales"]
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.20, 
                                                    random_state=144)
lm = LinearRegression() 
model = lm.fit(X_train, y_train)
```

```python
np.sqrt(mean_squared_error(y_train, model.predict(X_train)))         # MSE train grubu
```

```python
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))           # MSE test grubu
```

```python
model.score(X_train, y_train)
```

* Ã¼stteki 4 kod bloÄŸu iÃ§in :
* verinin tamamÄ± train ve test olmak Ã¼zere farklÄ± oranlarda (80-20 gibi) ikiye ayrÄ±lÄ±yor. fakat burada bu iki grubun verinin hangi kÄ±sÄ±mlarÄ±ndan seÃ§ileceÄŸi Ã¶nem kazanÄ±yor. ilk kod satÄ±rÄ±ndaki random_state argÃ¼manÄ±na farklÄ± deÄŸerler girilirse train ve test gruplarÄ± farklÄ± ÅŸekillerde yani verinin farklÄ± kÄ±sÄ±mlarÄ±ndan seÃ§imiÅŸ olur. her deneme sonrasÄ± train ve test grubu iÃ§in MSE deÄŸerlerinin deÄŸiÅŸtiÄŸini gÃ¶zlemliyoruz.
* bu durumun Ã¶nÃ¼ne geÃ§mek iÃ§in cross validasyon iÅŸlemi yapÄ±lÄ±r

* aÅŸaÄŸÄ± kÄ±sÄ±m iÃ§in :
* cross_val_score ile rsquare deÄŸeri hesaplanÄ±yor. burada cv = 10 ile 10 farklÄ± rsquare hesaplanÄ±yor ve mean ile ortalamasÄ± alÄ±nÄ±yor.

```python
cross_val_score(model, X, y, cv = 10, scoring = "r2")       # 10 farklÄ± rsquared deÄŸeri bulunuyor
```

```python
cross_val_score(model, X, y, cv = 10, scoring = "r2").mean()
```

* 10 farklÄ± rsquare deÄŸerinin ortalamasÄ± alÄ±narak modelimiz iÃ§in daha gÃ¼venilir ve doÄŸrulanmÄ±ÅŸ rsquare deÄŸeri bulunmuÅŸ olur 

```python
cross_val_score(model, X_train, y_train, cv = 10, scoring = "r2").mean()   

# train grubu iÃ§in bulduÄŸumuz valide edilmiÅŸ yani daha gÃ¼venilir rsquare deÄŸerini yukarÄ±daki model.score(X_train, y_train) ile karÅŸÄ±laÅŸtÄ±ralÄ±m 
```

```python
np.sqrt(-cross_val_score(model, 
                X_train, 
                y_train, 
                cv = 10, 
                scoring = "neg_mean_squared_error"))
```

* validasyon ile 10 farklÄ± MSE deÄŸeri buluyoruz
* aÅŸaÄŸÄ±da ise ortalamasÄ± alÄ±narak valide edilmiÅŸ doÄŸrulanmÄ±ÅŸ daha gÃ¼venilir MSE train bulunuyor.

```python
np.sqrt(-cross_val_score(model, 
                X_train, 
                y_train, 
                cv = 10, 
                scoring = "neg_mean_squared_error")).mean()
```

```python
np.sqrt(-cross_val_score(model, 
                X_test, 
                y_test, 
                cv = 10, 
                scoring = "neg_mean_squared_error"))
```

```python
np.sqrt(-cross_val_score(model, 
                X_test, 
                y_test, 
                cv = 10, 
                scoring = "neg_mean_squared_error")).mean()
```

# PLS

* baÄŸÄ±msÄ±z deÄŸiÅŸkenlerin kombinasyonu Ã¼zerinden baÄŸÄ±mlÄ± deÄŸiÅŸken aÃ§Ä±klanmaya Ã§alÄ±ÅŸÄ±lÄ±r. 
* boyut indirgeme yaklaÅŸÄ±mÄ±nda P adet deÄŸiÅŸkenin taÅŸÄ±dÄ±ÄŸÄ± bilgiyi ondan daha az sayÄ±da C deÄŸiÅŸken ile ifade etmeye Ã§alÄ±ÅŸma 

### PLS - KÄ±smi En KÃ¼Ã§Ã¼k Kareler Regresyonu

* Ã‡ok boyutluluk laneti p > n 2 ne Ã§Ã¶zÃ¼m sunar
* Ã‡oklu doÄŸrusal baÄŸlantÄ± problemine Ã§Ã¶zÃ¼m sunar
* PLS de PCR gibi baÄŸÄ±msÄ±z deÄŸiÅŸkenlerin doÄŸrusal kombinasyonlarÄ±nÄ± bulur. Bu doÄŸrusal kombinasyonlar bileÅŸen yada latent deÄŸiÅŸken olarak adlandÄ±rÄ±lÄ±r.
* PLS NIPALS isimli algoritmanÄ±n Ã¶zel bir halidir, iteratif olarak baÄŸÄ±mlÄ± deÄŸiÅŸken ile yÃ¼ksek korelasyona sahip deÄŸiÅŸkenler arasÄ±ndaki gizil (latent) iliÅŸkiyi bulmaya Ã§alÄ±ÅŸÄ±r. Bu durum PLS ile PCR Ä± ayÄ±ran noktadÄ±r.

### PCR vs PLS
* PCR da doÄŸrusal kombinasyonlar yani bileÅŸenler baÄŸÄ±msÄ±z deÄŸiÅŸken uzayÄ±ndaki deÄŸiÅŸkenliÄŸi maksimum ÅŸekilde Ã¶zetleyecek ÅŸekilde oluÅŸturulur. BU durum baÄŸÄ±mlÄ± deÄŸiÅŸkeni aÃ§Ä±klama yeteneÄŸi olmamasÄ±na sepep olmakta.
* PLS te ise bileÅŸenler baÄŸÄ±mlÄ± deÄŸiÅŸken ile kovaryansÄ± maksimum ÅŸekild Ã¶zetleyecek ÅŸekilde oluÅŸturulur.
* PCR baÄŸÄ±msÄ±z deÄŸiÅŸkenlerin kendi arasÄ±nda PLS ise baÄŸÄ±mlÄ± deÄŸiÅŸkene odaklÄ± bir ÅŸekilde bu bileÅŸenleri oluÅŸturma iÅŸlemi yapÄ±yor.
* PLS ve PCR da boyut indirgemesi yapÄ±lÄ±yor. PLS te boyut indirgeme baÄŸÄ±mlÄ± deÄŸiÅŸken odaÄŸÄ±nda yapÄ±lÄ±yor bu nedenle gÃ¶zetimli boyut indirgeme prosedÃ¼rÃ¼ olarak, PCR da ise baÄŸÄ±mlÄ± deÄŸikenle bir iliÅŸki olmaksÄ±zÄ±n  baÄŸÄ±msÄ±z deÄŸiÅŸkenleri kendi iÃ§inde deÄŸerlendirdiÄŸimizden dolayÄ± gÃ¶zetimsiz boyut indirgeme prosedÃ¼rÃ¼ olarak gÃ¶rÃ¼lebilir.
* iki yÃ¶ntemde de bir tuning parametresi vardÄ±r ve oda bilÅŸene sayÄ±sÄ±dÄ±r. Optimum bileÅŸen sayÄ±sÄ±nÄ± belirlemek iÃ§in Cross Validation kullanÄ±lÄ±r.

# Ridge Regresyon

* problem burada aynÄ± aykÄ±rÄ± gÃ¶zlemlere duyarlÄ± olmasÄ± varsayÄ±m bozulumlarÄ± gibi problemler ile bu doÄŸrusal regresyon modelinin kabiliyetlerini kullanarak daha ileri ve baÅŸarÄ±lÄ± bir dÃ¼zeye nasÄ±l getiririz Ã§abasÄ± onucu ortaya Ã§Ä±kan modelleme tÃ¼rÃ¼dÃ¼r. .

* L2 : dÃ¼zenlelileÅŸtirme yÃ¶ntemi olarak ifade ediliyor. her SSE ile elde edilen katsayÄ±lara ceza uygulamasÄ± yapÄ±lÄ±r. lambda ile verilecek cezayÄ± bazÄ± yÃ¶ntemlerle optimize ederek ayarlama imkanÄ±na sahip oluyoruz. 

## Model

```python
# train / test grubu ayÄ±rma
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
ms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import Ridge
```

```python
# model kurma fit etme
ridge_model = Ridge(alpha = 0.1).fit(X_train, y_train)   # alpha parametresi lambda yÄ± ifade eder
```

```python
ridge_model
```

```python
# katsayÄ±lar
ridge_model.coef_
```

```python
# aÅŸaÄŸÄ±daki dÃ¶ngÃ¼de kullanÄ±lacak lambda deÄŸerleri
10**np.linspace(10,-2,100)*0.5 
```

```python
# farklÄ± alpha (lambda) deÄŸerleri iÃ§in oluÅŸacak KATSAYI deÄŸerlerini gÃ¶rmek iÃ§in kurulan dÃ¶ngÃ¼

lambdalar = 10**np.linspace(10,-2,100)*0.5 

ridge_model = Ridge()                              # model nesnesi oluÅŸturuyoruz
katsayilar = []                                    # boÅŸ katsayÄ±lar listesi olluÅŸturuyoruz

for i in lambdalar:
    ridge_model.set_params(alpha = i)              # her iterasyonda alpha deÄŸeri iÃ§in parametreler oluÅŸturulacak
    ridge_model.fit(X_train, y_train)              # verilen parametreler ile model fit edilecek
    katsayilar.append(ridge_model.coef_)           # modelden elde edilen katsayÄ±lar, katsayÄ±lar listesine append ediliyor
    

    
ax = plt.gca()
ax.plot(lambdalar, katsayilar) 
ax.set_xscale('log') 

plt.xlabel('Lambda(Alpha) DeÄŸerleri')
plt.ylabel('KatsayÄ±lar/AÄŸÄ±rlÄ±klar')
plt.title('DÃ¼zenlileÅŸtirmenin Bir Fonksiyonu Olarak Ridge KatsayÄ±larÄ±');
```

* lambda nÄ±n alacaÄŸÄ± herbir deÄŸer iÃ§in katsayÄ±larÄ±n alacaÄŸÄ± deÄŸerler gÃ¶zlemleniyor. 
* ridge regresyonun lambda muamemelesi katsayÄ±larÄ± hiÃ§birzaman sÄ±fÄ±r yapmamak ama tÃ¼m katsayÄ±larÄ± modelde tutup bu katsayÄ±larÄ±n modele olan etkilerini ayarlayabilmek sÄ±fÄ±ra yaklaÅŸtÄ±rmak mesela.
* burada katsayÄ±lara belirlenecek lambda deÄŸerine gÃ¶re Ã¶yle bir ceza uyguluyoruz ki  bu katsayÄ±lar bazen etkilerini Ã¶nemlerine gÃ¶re kaybetmiÅŸ oluyor. secilecek lambda deÄŸerine gÃ¶re deÄŸiÅŸkenlerin gÃ¶rece etkilerini tahmin fonksiyonundaki etkilerini ayarlayabilmiÅŸ oluyoruz ama diyoruz ki yinede bunlar modelde kalsÄ±n sÄ±fÄ±ra yaklaÅŸtÄ±ralÄ±m ama komple modelden Ã§Ä±karmayalÄ±m demiÅŸ oluyoruz ridge kullanÄ±ldÄ±ÄŸÄ±nda.

## Tahmin

# Lasso Regresyon

* ridge model ile aynÄ± tanÄ±m geÃ§erli, aralarÄ±ndaki fark lassoda katsayÄ±larÄ± 0 yapacak cezalar uygulanÄ±r. BU ÅŸekilde deÄŸiÅŸken seÃ§imi yapar

* formule dikkat edilirse Ridge formulu ile aynÄ± sadece ceza terimi mutlak deÄŸer iÃ§erisinde, lambda ayar parametremiz yine mevcut

## Model

* model adÄ±mlarÄ± ridge ile aynÄ± sorun olursa ordaki kod aÃ§Ä±klamalarÄ±na bakÄ±labilir

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
ms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha = 0.1).fit(X_train, y_train)
```

```python
lasso_model
```

```python
lasso_model.coef_
```

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5

lasso = Lasso()
katsayilar = []

for i in lambdalar:
    lasso.set_params(alpha=i)
    lasso.fit(X_train, y_train)
    katsayilar.append(lasso.coef_)
    
ax = plt.gca()
ax.plot(lambdalar*2, katsayilar)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

* buradaki fark ÅŸudur. ridge regresyonda lambda deÄŸerine gÃ¶re katsayÄ±lar 0 a yaklaÅŸÄ±yordu, lasso da katsayÄ±larÄ±n bir kÄ±smÄ± 0 oluyor

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
ms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)


```

```python
from sklearn.linear_model import ElasticNet
```

```python
enet_model = ElasticNet().fit(X_train, y_train)
```

```python
enet_model.coef_
```

```python
enet_model.intercept_
```

## Model Tuning

```python
from sklearn.linear_model import ElasticNetCV
```

```python
enet_cv_model = ElasticNetCV(cv = 10, random_state = 0).fit(X_train, y_train)
```

```python
enet_cv_model.alpha_
```

```python
enet_cv_model
```

```python
enet_tuned = ElasticNet(alpha = enet_cv_model.alpha_).fit(X_train,y_train)
```

```python
y_pred = enet_tuned.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```
